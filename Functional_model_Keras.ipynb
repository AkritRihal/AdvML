FUNCTIONAL MODELS

Sequential models are not always the solution, sometimes one or more models need conctetnation , separartion and lot more.

To enable a world of such endless possibilities we use Functional models

```python
from keras.models import Model
```

```python
model = Model(inputs = x , outputs = [output1,output2])    #tell which layer or layers are input and which are output
```

```python
from keras.layers import *
```

```python
x = Input(shape=(3,)) #shape of the input to the input layer
```

here we need to give name to each layer alongwith the the laye rfrom which it is taking the input or previous layer

```python
hidden1 = Dense(128,activation = 'relu')(x)
hidden2 = Dense(64, activation = 'relu' )(hidden1)

output1 = Dense(1,activation = 'linear')(hidden2)
output2 = Dense(1, activation='sigmoid')(hidden2)
```

```python
model.summary()
```

```python
from keras.utils import plot_model
plot_model(model, show_shapes=True)
```
